```python
NEW_VARIABLE = ['SalaryPerHour', 'Employer Provided Salary', 'Minimum Salary', 'Maximum Salary', 'Average Salary', 'State', 'Company_Age_years']
DATA_PATH ='./data/Raw_Data/rawData.csv'
CLEAN_DATA_PATH = './data/Cleaned_Data/cleanData.pkl'
```

# First Import Our Data

```python
# import important packages
import pandas as pd
import numpy as np
import pickle
Data = pd.read_csv(DATA_PATH)
Data.head(10)
```

# Data Wrangling

## First Data Cleaning 

**1- Salary Estimate Variable**

I notice that salary varaible has (Glassdoor est.) on it at first so I need to delete it.

```python
Data['Salary Estimate'] = Data['Salary Estimate'].apply(lambda x : x.split('(')[0])
Data['Salary Estimate'].value_counts()
```

notice that there is a salary with -1 so i decided to ignore this Observations.

```python
Data = Data[Data['Salary Estimate'] != '-1']
Data['Salary Estimate'].value_counts()
```

need to remove $ sign and K sign.

```python
Data['Salary Estimate'] = Data['Salary Estimate'].apply(lambda x: x.replace('$', '').replace('K', '000'))
Data['Salary Estimate'].value_counts()
```

notice also there is 'Per Hour and Employed Provided Salary' for that i will creat new columns for those.

```python
Data[NEW_VARIABLE[0]] = Data['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
Data[NEW_VARIABLE[1]] = Data['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary:' in x.lower() else 0)
Data
```

```python
Data['Salary Estimate'] = Data['Salary Estimate'].apply(lambda x: x.lower().replace('per hour', '').replace('employer provided salary:', ''))
```

create new to columns for min and max Salary 

```python
Data[NEW_VARIABLE[2]] = Data['Salary Estimate'].apply(lambda x: x.split('-')[0]).astype(int)
Data[NEW_VARIABLE[3]] = Data['Salary Estimate'].apply(lambda x: x.split('-')[1]).astype(int)
Data[NEW_VARIABLE[4]] = (Data['Maximum Salary'] + Data['Minimum Salary']) / 2

Data.info()
```

Now I finished cleaning Salary Column.

**2- Caompany Name Variable**

Remove rating from company names

```python
Data['Company Name'] = Data['Company Name'].apply(lambda x :x.split('\n')[0])

Data
```

**3- Location**

create new columns for job states

```python
Data[NEW_VARIABLE[5]] = Data['Location'].apply(lambda x: x.split(',')[1])
Data.State.value_counts()
```

**4-Transformed founded date into age of company**

```python
Data[NEW_VARIABLE[6]] = Data['Founded'].apply(lambda x : x if x < 0 else 2021 - x)
Data
```

**5- Job Describtion**

```python

#python
Data['python'] = Data['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
 
#r studio 
Data['R'] = Data['Job Description'].apply(lambda x: 1 if 'r studio' in x.lower() or 'r-studio' in x.lower() else 0)

#spark 
Data['spark'] = Data['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)

#aws 
Data['AWS'] = Data['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)

#excel
Data['excel'] = Data['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
```

## Save Cleaned Data

```python
Data.to_pickle(CLEAN_DATA_PATH)
```
